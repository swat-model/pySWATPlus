```python
import pandas as pd
import numpy as np
from pySWATPlus.TxtinoutReader import TxtinoutReader
from pySWATPlus.FileReader import FileReader
```

```python
txtinout_folder = 'path/to/txtinout'
filename =  'plants.plt'
```

```python
#The folder pointed by txtinout_folder must have a .exe file to run the simulation
reader = TxtinoutReader(txtinout_folder)
```

```python
#Read plants.plt file
plants_reader = reader.register_file(filename, 
                    has_units=False,
                    index='name')
```

```python
plants_reader.header_file
```

```python
plants_reader.df
```

```python
reader.set_beginning_and_end_year(2009, 2011)
```

```python
reader.set_warmup(1)
```

```python
#Enable the 'channel_sd' in the print.prt file (the other ones are mantained as they are)
reader.enable_object_in_print_prt(obj='channel_sd', daily=True, monthly=False, yearly=False, avann=False)
```

```python
#Run the model with the default settings
reader.run_swat(show_output=True)
```

```python
#Read the results
channel_sd = reader.register_file('channel_sd_day.txt', 
                    has_units=True)
channel_sd.df
```

```python
#For modifying the input files, either modify the stored dataframe and save it, and then run the model again
plants_reader.df.loc['agrc', 'bm_e'] = 40
plants_reader.df.head()
```

```python
plants_reader.overwrite_file()
```

```python
#Or directly the model specifying the modifications
txtinout_path = reader.run_swat(params = {'plants.plt': ('name', [('agrc', 'bm_e', 40)])}, show_output=True)
```

```python
#Or you can also run multiple simulations on parallel
txtinout_paths = reader.run_parallel_swat(params = [{'plants.plt': ('name', [('bana', 'bm_e', 45)])}, {'plants.plt': ('name', [('bana', 'bm_e', 40)])}], n_workers = 2, parallelization='processes')
txtinout_paths
```
